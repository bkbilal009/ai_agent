```python
from datetime import datetime
from zoneinfo import ZoneInfo

def get_current_time(city):
    city_timezones = {
        "lahore": "Asia/Karachi",
        "peshawar": "Asia/Karachi",
        "new york": "America/New_York"
    }

    city = city.lower()

    if city not in city_timezones:
        return "Sorry, I do not support this city yet."

    timezone = ZoneInfo(city_timezones[city])
    current_time = datetime.now(timezone)

    return f"Current time in {city.title()} is {current_time.strftime('%H:%M:%S')}"
```

```python
def time_agent(user_query):
    user_query = user_query.lower()

    if "lahore" in user_query:
        return get_current_time("lahore")

    elif "peshawar" in user_query:
        return get_current_time("peshawar")

    elif "new york" in user_query:
        return get_current_time("new york")

    else:
        return "Please ask like. What time in Lahore?"
```

```python
print(time_agent("What time in Lahore?"))
print(time_agent("What time in Peshawar?"))
print(time_agent("What time in New York?"))
```

```python
while True:
    query = input("Ask time or type exit: ")

    if query.lower() == "exit":
        print("Agent stopped")
        break

    response = time_agent(query)
    print(response)
```
